# **Implementation for is_null()**

```python
def my_is_null(df , null_vals = []):


    '''
    Check for missing values in a DataFrame.

    Parameters:
    df (DataFrame)      : The input DataFrame.
    null_vals (list)    : list to pass values ​​that are considered as null values.

    Returns:
    DataFrame: A DataFrame indicating the presence of missing values.
               The same shape as the input DataFrame with boolean values.
    '''




    '''
    Initialize a DataFrame to store missing values indicators (true if it is null) default = true
    df.index    ----- > RangeIndex(start=0, stop=500, step=1)
    df.columns  ------> Index(['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership', 'Yearly Amount Spent'],dtype='object')
    '''
    missing_values = pd.DataFrame ( index = df.index , columns = df.columns , dtype=bool )


    '''
    Iterate over each column and row of the DataFrame
    Iterate over each row in a column in  the DataFrame
    Check if the value is null (NaN)

    '''

    for col in df.columns:
        for idx in df.index:
            if pd.isna(df[col][idx]) or df[col][idx] in null_vals:
                missing_values.at[idx, col] = True
            else:
                missing_values.at[idx, col] = False

    return missing_values
```

# **Implementation for is_null().sum()**

```python
def my_missing_values_col_sum(missing_values_df):


    """
    Calculate the sum of missing values in each column of a DataFrame.

    Parameters:
    missing_values_df (DataFrame): A DataFrame indicating the presence of missing values.
                                   The same shape as the input DataFrame with boolean values.

    Returns:
    dict: A dictionary containing the column names as keys and their corresponding sum of missing values.
    """

    # Initialize an empty dictionary to store the column and  sums of nulls
    col_sums = {}


    '''
    Iterate over each column of the input DataFrame
    Initialize variable to store the sums of null in each column
    Iterate over each row in the column of the input DataFrame
    check if value equal true
    add 1 to col_sum
    Store the column name and its sum in the dictionary
    '''
    for col in missing_values_df.columns:
      col_sum = 0
      for value in missing_values_df[col]:
        if value:
          col_sum += 1
      col_sums[col] = col_sum

    # print the dictionary containing the column sums
    print("*********************** Sum Nulls in each Column *********************\n")
    for col, col_sum in col_sums.items():
      print(f"{col}: {col_sum}")
    print("\n*********************************************************************\n")



```
